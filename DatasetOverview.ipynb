```python
import os
import pandas as pd
import numpy as np
import pandas.api.types as ptypes

# Folder containing the CSV files
DATA_FOLDER = "./data/csv"

# ---------------------------------------------------------------------
# 1) Define for each CSV the specific columns (fields) we want to show:
# ---------------------------------------------------------------------
KEY_FIELDS = {
    "sonar_seaking_ros.csv": [
        "Time", 
        "header.seq", 
        "header.stamp.secs", 
        "header.stamp.nsecs", 
        "header.frame_id", 
        "intensities_0",
        "intensities_49",
    ],
    "sonar_micron_ros.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "intensities_0",
        "intensities_396"
    ],
    "depth_sensor.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "depth"
    ],
    "imu_adis.csv": [
        "Time", 
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "roll",
        "pitch",
        "yaw",
        "ax",
        "ay",
        "az",
        "gx",
        "gy",
        "gz"
    ],
    "odometry.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "child_frame_id",
        "pose.pose.position.x",
        "pose.pose.position.y",
        "pose.pose.position.z",
        "pose.pose.orientation.x",
        "pose.pose.orientation.y",
        "pose.pose.orientation.z",
        "pose.pose.orientation.w"
    ],
    "sonar_micron.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "beam_data_0",
        "beam_data_396"
    ],
    "imu_xsens_mti_ros.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "orientation.x",
        "orientation.y",
        "orientation.z",
        "orientation.w",
        "angular_velocity.x",
        "angular_velocity.y",
        "angular_velocity.z"
    ],
    "imu_xsens_mti.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "roll",
        "pitch",
        "yaw",
        "qx",
        "qy",
        "qz",
        "qw",
        "ax",
        "ay",
        "az",
        "gx",
        "gy",
        "gz"
    ],
    "dvl_linkquest.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "altitudeBeam_0",
        "altitudeBeam_3",
        "bottomVelocityBeam_0",
        "bottomVelocityBeam_3",
        "velocityEarth_0",
        "velocityEarth_1",
        "velocityEarth_2"
    ],
    "tf.csv": [
        "Time",
        # 'transforms' is a single column containing a list-like object (likely strings)
        "transforms"
    ],
    "sonar_seaking.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "angle_grad",
        "beam_data_0",
        "beam_data_49"
    ],
    "imu_adis_ros.csv": [
        "Time",
        "header.seq",
        "header.stamp.secs",
        "header.stamp.nsecs",
        "header.frame_id",
        "orientation.x",
        "orientation.y",
        "orientation.z",
        "orientation.w",
        "angular_velocity.x",
        "angular_velocity.y",
        "angular_velocity.z"
    ],
}

# ---------------------------------------------------------------------
# 2) Read each CSV and produce a per-file Markdown summary table
# ---------------------------------------------------------------------

# We'll gather CSV files in alphabetical order based on the keys
all_csvs = sorted(KEY_FIELDS.keys())

for csv_file in all_csvs:
    fpath = os.path.join(DATA_FOLDER, csv_file)
    if not os.path.exists(fpath):
        print(f"\n**Skipping** {csv_file} because it does not exist in {DATA_FOLDER}.\n")
        continue
    
    # Read the CSV
    df = pd.read_csv(fpath)
    rows, cols = df.shape
    dimension_str = f"{rows} x {cols}"

    # Extract first and last row as dict
    first_row = df.iloc[0].to_dict()
    last_row = df.iloc[-1].to_dict()

    # Print a heading for this file
    print(f"\n#### File: `{csv_file}`")
    print()
    print(f"**Dimension**: {dimension_str}  ")
    print()

    # Build a table of key fields (First row, Last row, Min, Max)
    table_rows = []
    fields = KEY_FIELDS[csv_file]
    for field in fields:
        # First/Last values:
        first_val = first_row.get(field, "N/A")
        last_val = last_row.get(field, "N/A")

        # Min/Max (only if it's a numeric column)
        if field in df.columns and ptypes.is_numeric_dtype(df[field]):
            col_min = df[field].min()
            col_max = df[field].max()
        else:
            col_min = "N/A"
            col_max = "N/A"

        table_rows.append((field, first_val, last_val, col_min, col_max))

    # Print as a Markdown table with columns: Field, First, Last, Min, Max
    print("| **Field** | **First Row Value** | **Last Row Value** | **Min Value** | **Max Value** |")
    print("|-----------|----------------------|---------------------|---------------|---------------|")
    for (field, val1, val2, min_val, max_val) in table_rows:
        print(f"| **{field}** | {val1} | {val2} | {min_val} | {max_val} |")

    print()  # Blank line after each table
```

```python
import pandas as pd
import os
from IPython.display import display

# Define a function to preview the first and last row of a CSV file
def preview_csv_rows(file_path):
    try:
        df = pd.read_csv(file_path)
        first_row = df.iloc[0]
        last_row = df.iloc[-1]

        print(f"\nFile: {os.path.basename(file_path)}")
        print("First Row:")
        display(first_row)
        print("\nLast Row:")
        display(last_row)
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")

# Directory containing CSV files
data_directory = "data/csv"

# Process one or two CSV files to preview
csv_files = [f for f in os.listdir(data_directory) if f.endswith(".csv")]

# Preview the first and last rows of up to two files
for file_name in csv_files:
    file_path = os.path.join(data_directory, file_name)
    preview_csv_rows(file_path)
```

```python
import pandas as pd
import os
from IPython.display import Markdown, display

# Define a function to extract key metadata from a CSV file
def extract_key_info(file_path):
    try:
        df = pd.read_csv(file_path)
        dimensions = df.shape
        first_row = df.iloc[0]
        last_row = df.iloc[-1]

        # Define key columns based on the file name
        file_name = os.path.basename(file_path)
        if "depth_sensor" in file_name:
            key_columns = ["Time", "depth"]
        elif "dvl_linkquest" in file_name:
            key_columns = ["Time", "bottomVelocityBeam_0", "altitudeBeam_0"]
        elif "imu_adis" in file_name or "imu_xsens_mti" in file_name:
            key_columns = ["Time", "roll", "pitch", "yaw", "ax", "ay", "az"]
        elif "odometry" in file_name:
            key_columns = ["Time", "pose.pose.position.x", "pose.pose.position.y", "pose.pose.position.z"]
        elif "sonar_micron" in file_name or "sonar_seaking" in file_name:
            key_columns = ["Time", "beam_data_0", "beam_data_1"]
        elif "tf" in file_name:
            key_columns = ["Time", "transforms"]
        else:
            key_columns = df.columns[:5].tolist()  # Default to the first 5 columns

        # Extract key information from the rows
        first_values = {col: first_row.get(col, "N/A") for col in key_columns}
        last_values = {col: last_row.get(col, "N/A") for col in key_columns}

        return {
            "File Name": file_name,
            "Dimensions": f"{dimensions[0]} x {dimensions[1]}",
            "Key Columns": key_columns,
            "First Row": first_values,
            "Last Row": last_values
        }
    except Exception as e:
        return {
            "File Name": os.path.basename(file_path),
            "Error": str(e)
        }

# Generate Markdown table for a single file's metadata
def generate_markdown_table(metadata):
    markdown_table = f"### {metadata['File Name']}\n\n"
    markdown_table += "| Key Field | First Value | Last Value |\n"
    markdown_table += "|-----------|-------------|------------|\n"

    for col in metadata["Key Columns"]:
        first_val = metadata["First Row"].get(col, "N/A")
        last_val = metadata["Last Row"].get(col, "N/A")
        markdown_table += f"| {col} | {first_val} | {last_val} |\n"

    return markdown_table

# Directory containing CSV files
data_directory = "data/csv"

# Process all CSV files in the directory
markdown_output = ""
for file_name in os.listdir(data_directory):
    if file_name.endswith(".csv"):
        file_path = os.path.join(data_directory, file_name)
        metadata = extract_key_info(file_path)
        markdown_output += generate_markdown_table(metadata) + "\n\n"

# Display Markdown output
def display_markdown_table(markdown_output):
    display(Markdown(markdown_output))

display_markdown_table(markdown_output)
```

```python
import pandas as pd
import os
from IPython.display import Markdown, display

# Define a function to extract metadata from a CSV file
def extract_csv_metadata(file_path):
    try:
        df = pd.read_csv(file_path)
        dimensions = df.shape
        first_value = df.iloc[0].to_dict()
        last_value = df.iloc[-1].to_dict()

        return {
            'File Name': os.path.basename(file_path),
            'Dimensions': f"{dimensions[0]} x {dimensions[1]}",
            'First Value': first_value,
            'End Value': last_value
        }
    except Exception as e:
        return {
            'File Name': os.path.basename(file_path),
            'Error': str(e)
        }

# Generate Markdown table for metadata
def generate_markdown_table(metadata_list):
    markdown_table = "| Data | Dimension | First Value | End Value |\n"
    markdown_table += "|------|-----------|-------------|-----------|\n"
    
    for metadata in metadata_list:
        first_value = metadata.get('First Value', 'N/A')
        end_value = metadata.get('End Value', 'N/A')
        markdown_table += f"| {metadata['File Name']} | {metadata.get('Dimensions', 'N/A')} | {first_value} | {end_value} |\n"

    return markdown_table

# Directory containing CSV files
data_directory = "data/csv"

# Process all CSV files in the directory
metadata_list = []
for file_name in os.listdir(data_directory):
    if file_name.endswith(".csv"):
        file_path = os.path.join(data_directory, file_name)
        metadata = extract_csv_metadata(file_path)
        metadata_list.append(metadata)

# Generate Markdown table
markdown_output = generate_markdown_table(metadata_list)

# Display Markdown output
def display_markdown_table(markdown_output):
    display(Markdown(markdown_output))

display_markdown_table(markdown_output)
```
